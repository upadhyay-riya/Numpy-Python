### 1. Data Analysis: Calculating Descriptive Statistics

#### Problem:
You have a dataset of daily temperatures recorded over a month, and you need to calculate the mean, median, and standard deviation of the temperatures to understand the overall trend and variability.

```python
import numpy as np

# Example dataset: daily temperatures for a month (in Celsius)
temperatures = np.array([22, 21, 23, 20, 19, 24, 25, 26, 24, 23, 22, 21, 20, 19, 18, 17, 16, 18, 19, 20, 22, 21, 23, 25, 26, 27, 28, 29, 30, 31])

# Calculate mean, median, and standard deviation
mean_temp = np.mean(temperatures)
median_temp = np.median(temperatures)
std_temp = np.std(temperatures)

print(f"Mean Temperature: {mean_temp:.2f}°C")
print(f"Median Temperature: {median_temp:.2f}°C")
print(f"Standard Deviation: {std_temp:.2f}°C")
```

### 2. Financial Analysis: Portfolio Return Calculation

#### Problem:
You have the historical returns of three different stocks and you want to calculate the overall portfolio return, assuming equal weights for each stock in the portfolio.

```python
# Historical returns of three stocks (in percentages)
stock_returns = np.array([
    [0.02, 0.03, -0.01, 0.04],
    [0.01, 0.05, -0.02, 0.03],
    [-0.01, 0.04, 0.01, 0.02]
])

# Calculate mean return for each stock
mean_returns = np.mean(stock_returns, axis=1)

# Assuming equal weights for each stock in the portfolio
weights = np.array([1/3, 1/3, 1/3])

# Calculate overall portfolio return
portfolio_return = np.dot(weights, mean_returns)

print(f"Portfolio Return: {portfolio_return:.2f}%")
```

### 3. Image Processing: Grayscale Image Transformation

#### Problem:
You have a color image and you need to convert it to a grayscale image to simplify further image processing tasks.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load a color image using OpenCV
image = cv2.imread('cat-in-tech.jpeg')

# Convert the image from BGR (OpenCV format) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale using the formula: 
# grayscale = 0.299*R + 0.587*G + 0.114*B
grayscale_image = 0.299 * image_rgb[:,:,0] + 0.587 * image_rgb[:,:,1] + 0.114 * image_rgb[:,:,2]

# Display the original and grayscale images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Grayscale Image')
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')

plt.show()
```

### 4. Linear Algebra: Solving a System of Linear Equations

#### Problem:
You have a system of linear equations and you need to find the values of the variables that satisfy all equations.

Example system of equations:
- \[ 2x + 3y = 8 \]
- \[ 3x + y = 5 \]

```python
# Coefficients matrix
A = np.array([
    [2, 3],
    [3, 1]
])

# Constants vector
B = np.array([8, 5])

# Solve the system of equations
solution = np.linalg.solve(A, B)

print(f"Solution: x = {solution[0]:.2f}, y = {solution[1]:.2f}")
```

### 5. Signal Processing: Fourier Transform

#### Problem:
You have a time series signal and you need to analyze its frequency components using the Fast Fourier Transform (FFT).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal: sine wave with added noise
sampling_rate = 1000
t = np.linspace(0, 1, sampling_rate)
frequency = 5
signal = np.sin(2 * np.pi * frequency * t) + 0.5 * np.random.randn(t.size)

# Perform Fourier Transform
fft_signal = np.fft.fft(signal)
fft_freq = np.fft.fftfreq(t.size, d=1/sampling_rate)

# Plot the original signal and its frequency components
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Signal')
plt.plot(t, signal)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(1, 2, 2)
plt.title('Frequency Components')
plt.stem(fft_freq, np.abs(fft_signal), 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.xlim(0, 20)  # Limit x-axis for better visualization

plt.tight_layout()
plt.show()
```
